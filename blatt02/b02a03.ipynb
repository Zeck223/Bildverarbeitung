```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show_image(image):
    """Helper function to display an image."""
    plt.imshow(image, cmap="gray")
    plt.axis("off")
```

## 3.1

100x100 Pixel mit 0 initialisiert und dtype=float

```python
png = np.full((100, 100), 0, dtype=float)
show_image(png)
```

Die erste Zeile abhängig von y-Wert auf `np.sin(0.1*y)` setzen.

```python
png[0] = np.sin(0.1 * np.arange(100))
show_image(png)
```

Erste Zeile kopieren, sodass alle gleich der ersten sind.

```python
png[1:] = png[0]
show_image(png)
```

## 3.2

```python
png = png + 1
png = png / 2
png = png * 255
plt.plot(png[0])
```

## 3.3

```python
png[0] = np.sin(1 * np.arange(100))
png[1:] = png[0]
show_image(png)
```

Die Erhöhung der Frequenz auf 1 sorgt dafür, dass die sanften Übergänge (bei 0.1) nun sehr ungenau dargestellt werden (Unterabtastung).
