```python
import matplotlib.pyplot as plt
import numpy as np
import skimage
```

```python
def show_image(image):
    """Helper function to display an image."""
    plt.imshow(image, cmap="gray")
    plt.axis("off")
    plt.show()
```

## 5.1

```python
def add_gaussian_noise(
        image: np.ndarray,
        mean: float,
        sigma: float
    ) -> np.ndarray:
    """
    Adds gaussian noise to the given image based on given mean and sigma.

    Args:
    image: Input image
    mean: Mean of the Gaussian noise
    sigma: Standard deviation of the Gaussian noise

    Returns:
        Image with gaussian noise added
    """
    noise = np.random.normal(mean, sigma, image.shape)

    noisy_image = image + noise

    if image.max() <= 1:
        noisy_image = np.clip(noisy_image, 0, 1)
    else:
        noisy_image = np.clip(noisy_image, 0, 255)

    return noisy_image
```

```python
mandrill_png = skimage.io.imread('mandrill.png')
show_image(mandrill_png)
mandrill_noisy = add_gaussian_noise(mandrill_png, mean=0, sigma=50)
show_image(mandrill_noisy)
```

## 5.2

```python
def add_salt_and_pepper_noise(
        image: np.ndarray,
        salt_or_pepper_prob: float
    ) -> np.ndarray:
    """
    Adds salt-and-pepper noise to the given image based on salt and pepper probabilities.

    Args:
        image: Input image
        salt_or_pepper_prob: Probability of a pixel being either salt or pepper noise

    Returns:
        Image with salt-and-pepper noise added.
    """
    noisy_image = np.copy(image)
    random_matrix = np.random.rand(*image.shape)
    noisy_image[random_matrix < salt_or_pepper_prob / 2] = 0
    noisy_image[random_matrix > 1 - salt_or_pepper_prob / 2] = 255
    return noisy_image
```

```python
salt_or_pepper_prob = 0.01

mandrill_noisy_sp = add_salt_and_pepper_noise(mandrill_png, salt_or_pepper_prob)
show_image(mandrill_noisy_sp)
```
