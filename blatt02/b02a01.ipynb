```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread, imsave
```

Kleine Hilfsfunktion :)

```python
def show_image(image):
    """Helper function to display an image."""
    plt.imshow(image, cmap="gray")
    plt.axis("off")
    plt.show()
```

## 1.1
Bild einlagen und mit `matplotlib` anzeigen.

```python
cat_png = imread('cat.png')
show_image(cat_png)
```

Ausschnitt des rechten Auges erzeugen, anzeigen und speichern unter `cat_right_eye.png`.

```python
cat_png_right_eye = cat_png[70:160, 120:220]
show_image(cat_png_right_eye)
imsave('cat_right_eye.png', cat_png_right_eye)
```

## 1.2

Negativ von `cat.png` erzeugen und anzeigen.
(Wenn man nur das Negativ anzeigen lassen möchte, dann könnte auch die Colormap auf `cmap="gray_r"` gesetzt werden.)

```python
cat_png_negative = 255 - cat_png
show_image(cat_png_negative)
```

## 1.3

a) Neues binär Bild (alle Pixel 0), gleiche größe wie `cat.png`.

```python
mask = np.full(cat_png.shape, 0)
```

b) Pixel im Bereich des rechten Auges auf 1 setzen.

```python
mask[70:160, 120:220] = 1
show_image(mask)
```

c) Anwendung der Maske auf das Originalbild.

```python
masked_cat = cat_png * mask
show_image(masked_cat)
```
