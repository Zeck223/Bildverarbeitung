```python
import math
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
```

```python
imgs = [imread(f"bildverbesserung/bild{idx}.png") for idx in range(1, 4)]
```

## 1.1
Durch die Transformation:

$$g = (f - f.min) / (f.max - f.min)$$

Wird die Intensität auf das Intervall [0, 1] transformiert, wodurch eine bessere Verteilung des möglichen Wertebereichs erreicht wird. Dadurch erhöht sich der Kontrast.

```python
img_1 = imgs[0].copy()
img_1 = (img_1 - img_1.min()) / (img_1.max() - img_1.min())
plt.imshow(img_1, cmap="gray", vmin=0, vmax=1)
```

## 1.2

Mit der folgenden Transormation werden, die Bildbereiche heller dargestellt, die unter einem bestimmten Schwellenwert liegen. In diesem Fall ist der Schwellenwert 50.

```python
img_2 = imgs[1].copy()
img_2 = np.where(img_2 < 50, img_2 + 50, img_2)
plt.imshow(img_2, cmap="gray", vmin=0, vmax=255)
```

Die Vorderfront des Gebäudes hat 12 Fenster.

## 1.3

Die Transformation setzt alle Bildwerte, die eine Intensität größer als 100 haben auf 255. Dadurch wird die im Original graue Straße nun weiß dargestellt (Die kleinen Wege auch). Da aber der Hauptteil des Bildes sehr dunkel ist, findet dort keine Veränderung statt.

```python
img_3 = imgs[2].copy()
img_3 = np.where(img_3 > 100, 255, img_3)
plt.imshow(img_3, cmap="gray", vmin=0, vmax=255)
```
