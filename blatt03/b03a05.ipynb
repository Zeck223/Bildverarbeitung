# Aufgabe 5: Überlaufbehandlung bei Bildern

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
```

```python
einstein_png = imread("einstein.png")
plt.title("Originalbild")
plt.imshow(einstein_png, cmap="gray", vmin=0, vmax=255)
```

## 5.1
Jedes Pixel um 100 Graustufen aufhellen mit wrap-around.

```python
def brighter_with_wrap_around(img):
    return img + 100

plt.title("Brighter with wrap-around")
plt.imshow(
    brighter_with_wrap_around(einstein_png),
    cmap="gray",
    vmin=0,
    vmax=255
)
```

## 5.2

```python
def brighter_with_saturation(img):
    # Kovertiere das Bild in einen Datentyp, der Überläufe zulässt
    img = img.astype(np.int32)
    return np.clip(img + 100, a_min=0, a_max=255)

plt.title("Brighter with saturation")
plt.imshow(
    brighter_with_saturation(einstein_png),
    cmap="gray",
    vmin=0,
    vmax=255
)
```

## 5.3

```python
def brighter_with_rescaling(img):
    tmp = img.astype(np.int32) + 100
    tmp = (tmp - tmp.min()) / (tmp.max() - tmp.min()) * 255
    return tmp.astype(np.uint8)

plt.title("Brighter with rescaling")
plt.imshow(
    brighter_with_rescaling(einstein_png),
    cmap="gray",
    vmin=0,
    vmax=255
)
plt.show()
```
