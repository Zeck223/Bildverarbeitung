```python
# 3.1
l = list(range(100, 200, 3))
```

```python
# 3.2
def sum_of_elements(lst: list) -> float:
    """Returns the sum of all elements in the list.
    Args:
        lst: A list of numbers.
    Returns:
        The sum of all elements in the list.
    """
    result = 0
    for element in lst:
        result += element
    return result

sum_of_elements(l)
```

```python
# 3.3
def median(lst: list) -> float:
    """Returns the median of a list of numbers.
    
    Args:
        lst: A list of numbers.

    Returns:
        The median of the list.
    """
    sorted_lst = sorted(lst)
    n = len(sorted_lst)
    if n % 2 == 1:
        return sorted_lst[n // 2]
    else:
        return (sorted_lst[n // 2 - 1] + sorted_lst[n // 2]) / 2

median(l)
```

```python
# 3.4
def count_big_cross_sum(lst: list) -> int:
    """Returns the count of numbers in the list whose cross sum is greater than 5.
    
    Args:
        lst: A list of numbers.

    Returns:
        The count of numbers with a cross sum greater than 5.
    """
    count = 0
    for element in lst:
        cs = 0
        for digit in str(element):
            cs += int(digit)
        if cs > 5:
            count += 1
    return count

count_big_cross_sum(l)
```

```python
# 3.5
def count_primes(lst: list) -> int:
    """Returns the count of prime numbers in the list.
    
    Args:
        lst: A list of numbers.

    Returns:
        The count of prime numbers in the list.
    """
    count = 0
    for element in lst:
        if element < 2:
            continue
        is_prime = True
        for i in range(2, int(element**0.5) + 1):
            if element % i == 0:
                is_prime = False
                break
        if is_prime:
            count += 1
    return count

count_primes(l)
```
