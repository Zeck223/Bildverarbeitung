```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.exposure import equalize_hist, equalize_adapthist
from skimage.io import imread
```

```python
img = imread("moon.png")
plt.imshow(img, cmap="gray")
```

## 4.1 Histogram Equalization

```python
img_equalized = equalize_hist(img)
plt.imshow(img_equalized, cmap="gray")
```

**Problem:** Der globale Histogrammausgleich verbessert zwar den Gesamtkontrast, aber bei der Betrachtung an lokalen stellen ist der Kontrast zu stark.

## 4.2 Local Histogram Equalization

**Lösung:** Local Histogram Equalization teilt das Bild in Bereiche auf (hier in 8 Bereiche, da dies der Standardwert ist) und führt für jeden Bereich einen eigenen Histogrammausgleich durch. Dadurch werden bestimmte Bereicht nicht durch Ausreißer von anderen Bereichen beeinflusst. Der `clip_limit`erhöht den Kontrast (je höher, desto mehr Kontrast)

```python
# Adaptive Histogram Equalization (AHE)
img_ahe = equalize_adapthist(img, clip_limit=0.03)

# Visualisierung
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].imshow(img, cmap="gray")
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(img_equalized, cmap="gray")
axes[1].set_title("Global Histogram Equalization")
axes[1].axis("off")

axes[2].imshow(img_ahe, cmap="gray")
axes[2].set_title("Local Histogram Equalization")
axes[2].axis("off")

plt.tight_layout()
plt.show()
```
