Im Rahmen dieser Präsenzaufgabe soll eine Sinusfunktion manipuliert und so in verschiedene Sinusoide überführt werden.

```python
import matplotlib.pyplot as plt
from math import sin,cos,pi
import numpy as np
```

Hier zunächst ein Beispiel für das Plotten einer Funktion (verschobene Parabel) im Defintionsbereich (-4,4) und dem Wertebereich (-5,+5). Diesen Code könnt ihr als Ausgangspunkt für die weiteren Teilaufgaben verwenden.

```python
def f(x):
  return x**2-3

x = np.linspace(-4, 4, 1000)
y = list(map(f, x))

plt.plot(x,y)
plt.ylim((-5,5));
```

Definiert zunächst eine Funktion `f`, die nur den Sinus der Eingabe berechnet und plottet die Funktion an 1000 Stellen von `0` bis `10*pi`.




```python
def f(x):
  return sin(x)

x = np.linspace(0, 10*np.pi, 1000)
y = list(map(f, x))

plt.plot(x,y)
```

Nun soll die Funktion `f` so verändert werden, dass der Sinus um 2 Einheiten nach oben verschoben wird. Was muss verändert werden?

```python
def f(x):
  return sin(x) + 2

x = np.linspace(0, 10*np.pi, 1000)
y = list(map(f, x))

plt.plot(x,y)
```

Verändert nun die ursprüngliche Funktion `f` so, dass die Amplitude der Schwingungen dreimal so hoch ist wie vorher.

```python
def f(x):
  return 3 * sin(x)

x = np.linspace(0, 10*np.pi, 1000)
y = list(map(f, x))

plt.plot(x,y)
```

Verändert nun die ursprüngliche Funktion `f` so, dass die Frequenz 10-mal höher ist.

```python
def f(x):
  return sin(10 * x)

x = np.linspace(0, 10*np.pi, 1000)
y = list(map(f, x))

plt.plot(x,y)
```

Verändert nun die ursprüngliche Funktion `f` so, dass die Phase um `pi*0.2` verschoben wird. Was passiert, wenn die Phase um `pi` oder `pi/2` verschoben wird?

```python
def f(x):
  return sin(x-0.2*np.pi)

x = np.linspace(0, 10*np.pi, 1000)
y = list(map(f, x))

plt.plot(x,y)
```

Kombiniert nun zwei Sinusoiden zu einer Funktion. Der erste Sinusoid soll der normale Sinus sein, wohingegen der zweite Sinusoid ein Sinus mit einer 10-fach erhöhten Frequenz sein soll. Wie wird das Ergebnis aussehen?

```python
def f(x):
  return sin(x)

def g(x):
  return sin(10 * x)

x = np.linspace(0, 10*np.pi, 1000)
y = list(map(lambda x: g(f(x)), x))

plt.plot(x,y)
```

Erzeugt aus dieser Funktion nun eine Bildzeile, indem ihr die Funktion an 256 gleichmäßig verteilten Stellen von 0 bis `10*pi` abtastet und die Werte in ein Array schreibt. Ändert die Werte so, dass sie im Bereich 0 bis 255 liegen und ganzzahlig sind.

```python
line = np.array([g(f(x)) for x in np.linspace(0, 10*np.pi, 256)])

# Wertebereich von line in [0, 255] skalieren
line = (line - line.min()) / (line.max() - line.min()) * 255
```

Wiederholt die erzeugte Bildzeile nun 100 mal untereinander, um ein Bild der Größe 100x256 zu erzeugen. Zeigt das Bild an.

```python
img = np.tile(line, (100, 1))
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
```

Was passiert, wenn die Bildzeile nicht mehr aus 256 Pixeln besteht, sondern nur noch aus 50 Pixeln, aber weiterhin den Bereich von 0 bis `10*pi` abdecken soll?

```python
short_line = np.array([g(f(x)) for x in np.linspace(0, 10*np.pi, 50)])

short_line = (short_line - short_line.min()) / (short_line.max() - short_line.min()) * 255

short_line_img = np.tile(short_line, (100, 1))
plt.imshow(short_line_img, cmap="gray", vmin=0, vmax=255)
```

Verändert das Wellenmuster nun so, dass es sich nicht mehr von links nach rechts entlang der horizontalen Bildachse ausbreitet, sondern entlang der Diagonalen von unten links nach oben rechts. Das Bild soll dafür eine Größe von 256x256 haben.
